```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df=pd.read_csv('D:/data analysys/t1/farm5_lpn_2023-01-04_00-05.csv', names=['tag', 'index', 'w', 'x', 'y', 'z', 'p1', 's2', 's3','s4','h5','d6'])
```
